```python
table = [[y for y in x.strip()] for x in open("input_.txt","r").readlines()]
new_table = table.copy()
while True:
    total_occupied = 0
    for i in range(len(table)):
        for j in range(len(table[0])):
            occupied = 0
            for x in range(-1,2):
                for y in range(-1,2):
                    if not(x==0 and y==0) and (i+x)>=0 and (i+x)<len(table) and (j+y)>=0 and (j+y)<len(table[0]):
                        if table[i+x][j+y]=="#":
                            occupied+=1
            if table[i][j]=="L":
                new_table[i][j]="#"
            if table[i][j]=="#" and occupied >= 4:
                new_table[i][j]="L"  
            if new_table[i][j]=="#":
                total_occupied+=1
    if new_table==table:
        break
    table = new_table.copy()
print(total_occupied)
```
