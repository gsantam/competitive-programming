```python
import numpy as np
tile_dict = {}
tile_number = None
for line in open("input.txt").readlines():
    line = line.strip()
    if line.startswith("Tile"):
        if tile_number is not None:
            tile_dict[tile_number] = np.array(tile)
        tile_number = int(line.split(" ")[1][:-1])
        tile = []
    else:
        if len(line)>0:
            tile.append([x for x in line])
tile_dict[tile_number] = np.array(tile)
tile_list_dict = dict()
for key in tile_dict:
    if key not in tile_list_dict:
        tile_list_dict[key] = []
    for rot in range(4):
        for flip in range(3):
            if flip == 2:
                new_array = np.rot90(tile_dict[key].copy(),rot)
            else:
                new_array = np.flip(np.rot90(tile_dict[key].copy(),rot),axis = flip)
            hash_0 = hash(tuple(new_array[:,0].tolist()))
            hash_1 = hash(tuple(new_array[:,-1].tolist()))
            hash_2 = hash(tuple(new_array[0,:].tolist()))
            hash_3 = hash(tuple(new_array[-1,:].tolist()))

            tile_list_dict[key].append([hash_0,hash_1,hash_2,hash_3])
            
```
